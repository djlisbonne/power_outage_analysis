```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_inline.backend_inline import set_matplotlib_formats
from IPython.display import display, IFrame, HTML

from pathlib import Path

import plotly
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
```

```python
def import_data():
    df = pd.read_excel('outage.xlsx', skiprows=4)
    df.columns = df.iloc[0]
    df = df.drop(0)
    df = df.reset_index(drop=True)

    return df
```

```python
df = import_data()
```

```python
df.columns
```

```python
import plotly.graph_objects as go

# Create the figure with Plotly
fig = go.Figure()

# Add scatter plot to the figure
fig.add_trace(go.Scatter(
    x=df['OUTAGE.DURATION'], 
    y=df['ANOMALY.LEVEL'], 
    mode='markers', 
    marker=dict(color='blue', size=8),
    name='Power Outages'
))

# Add titles and labels
fig.update_layout(
    title='Relationship between OUTAGE.DURATION and ANOMALY.LEVEL',
    xaxis_title='Outage Duration (hours)',
    yaxis_title='Anomaly Level',
    template='plotly',
    height=600,
    width=800
)

# Show the interactive plot
fig.show()
```
